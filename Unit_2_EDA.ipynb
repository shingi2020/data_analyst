```python
#switch to installed mpg dataset
df2 = sns.load_dataset('mpg')
```

```python
df2
```

```python
df2['company'] = df2['name'].apply(lambda x: x.split()[0])
df2['model'] = df2['name'].apply(lambda x: ' '.join(x.split()[1:]))
df2.drop('name', axis=1, inplace=True)
```

```python
df2.head()
```

```python
df2.tail()
```

By spliting the column names, the analysis can be done by car company and car model, giving better insights into the data.

```python
sns.boxplot(df2['mpg'])
```

```python
sns.countplot(df2['company'])
```

```python
sns.countplot(df2['origin'])
```

In the next section, analysis of mpg of USA made cars and their model_years.

```python
#group all the USA made cars
model_usa =df2[df2['origin'] == 'usa']
```

```python
model_usa.groupby('model_year').describe()['mpg']
```

In the above summary, we are presented with mpg statistical summaries of cars that originated in the USA between 1970 - 1982. For instance, there are 22 cars in the dataset from 1970, with an average mpg of 15, a minimum of 9 mpg and a maximum of 22.

```python
plt.figure(figsize=(10,6))
plt.axhline(model_usa.mpg.mean(), color='r', linestyle='--', linewidth=2)
sns.boxplot(x='model_year', y='mpg', data=model_usa, palette='rainbow')
plt.show()
```
